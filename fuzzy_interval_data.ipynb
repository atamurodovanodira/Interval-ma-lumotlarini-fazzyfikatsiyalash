```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

```

```python
# Intervalli ma'lumotlarga misol yaratish

data = np.array([[2, 5, 8], [4, 7, 10], [6, 9, 11], [8, 11, 13]])
# Universal to‘plamni qurish
x = np.linspace(0, 15, 1000)

# Intervallar uchun noravshan mansubklik funksiyalarini yarating
membership_funcs = [fuzz.trimf(x, interval) for interval in data]

# Mansublik funksiyalarini chizing
for i, mf in enumerate(membership_funcs):
    plt.plot(x, mf, label=f'Interval {i+1}')

plt.title(' Interval ma\'lumotlarini fazzyfikatsiyalash')
plt.xlabel('Qiymati')
plt.ylabel('Mansubligi')
plt.legend()
plt.show()
```
